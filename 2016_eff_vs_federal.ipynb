```python
import pandas as pd
```

```python
eff = pd.read_excel('/content/drive/MyDrive/CSE291-B Project/2016 eff vs. federal/california_wt2_export_2016.xlsx')
fed = pd.read_excel('/content/drive/MyDrive/CSE291-B Project/2016 eff vs. federal/wiretap_b1_1231.2016.xlsx')
print(eff.dtypes)
print('-------------')
print(fed.dtypes)
eff['Judge'] = eff['Judge'].str.upper().str.strip()
eff['Prosecutor'] = eff['Prosecutor'].str.upper().str.strip()
eff['Offense'] = eff['Offense'].str.upper().str.strip()
fed['Judge'] = fed['Judge'].str.upper().str.strip()
fed['Prosecutor'] = fed['Prosecutor'].str.upper().str.strip()
fed['Offense'] = fed['Offense'].str.upper().str.strip()
```
